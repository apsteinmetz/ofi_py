## Data sampling and resampling

* We want to enrich our data set and we will see in the next class how we can combine data frames. 
* In this session we will just prepare the data. 
* Specifically we want to look at the **ICE BofAML US High Yield Master II Option-Adjusted Spread** (BAMLH0A0HYM2)
* We downloaded the data from the FRED database. To learn more about the dataset and FRED please check https://fred.stlouisfed.org/series/BAMLH0A0HYM2
    * FRED provides also an api that can be used to download the data. However, we will not explore this here.
    

* The data from FRED is daily and our data set is only available for quarter ends
* Therefore we need to resample the daily data to quarter ends

Use the following steps to achieve that:

### Step 1:
* Read the data into a new data frame **df_hy_oas**
* When reading the file **skip the first ten rows** since it contains meta data
* Set **observation_date** as index

<font color='red'>#ADD Your Code Here (one-three lines of code)</font>

```python
df_hy_oas = pd.read_excel("data/fred_hy.xls")[10:]
df_hy_oas.columns = ["date","oas"]
df_hy_oas['date'] = pd.to_datetime(df_hy_oas["date"])
df_hy_oas = df_hy_oas.set_index('date')
```

### Step 2
* We will first drop all NA values from the time series using *dropna()*
* Next we will convert the time series to basis points. The current time series is in percentage
* *This is done for you*

```python
df_hy_oas = df_hy_oas.dropna()
df_hy_oas.oas = df_hy_oas.oas * 100
df_hy_oas.head()
```

### Step 3:
* Use the *resample* method
* Change the frequency of the data to **quarterly** using the **last method**
* Store the data in a new data frame called **df_hy_oas_q**    

<font color='red'>#ADD Your Code Here (one-two lines of code)</font>

```python
df_hy_oas_q = df_hy_oas.resample('Q',convention='last').asfreq()
df_hy_oas_q.head()
```
