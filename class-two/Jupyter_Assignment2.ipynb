Setting NAs to zero is very dangerous. You file spurious correlations and lack of correlations when the data is simply missing. In a portfolio management context, you might imagine you have a diversified portfolio when you do not.

Correlations will change if we use asset levels instead of returns.

Let's look at complete cases only. Also let's keep the actual correlation values so we know that zero is zero and use a colormap that diverges with white being zero and vivid colors at the extremes.

```python
data_roa_corr_plot = data_asset_return_mat_corr.iloc[0:100, 0:100]
data_roa_corr_plot_dropna = data_roa_corr_plot.dropna()
data_roa_corr_dist_plot = data_roa_corr_plot_dropna
data_roa_corr_dist_plot
linkage = hc.linkage(data_roa_corr_dist_plot, method='average')
sns.clustermap(data_roa_corr_dist_plot, row_linkage=linkage, col_linkage=linkage,cmap="RdBu_r")
```

The banks shown above have pretty low asset change correlations. They happen to the first ones in the list, alphabetically. They also happen to be pretty small sized.  How does it look if we take the biggest banks?

We use `fillna(0)` rather than `dropna()` so that every bank is visible.  The analyst presenting this data must make sure that the audience understands that zero means zero AND missing data so no conclusions can be inferred.

```python
# get 100 biggest banks
biggest_banks = df_max_assets.head(50).index
df_biggest= df[df.financial_institution_name.isin(biggest_banks)]
df_biggest
df3= df_biggest[['financial_institution_name','date','total_assets']].pivot_table(index='date',columns='financial_institution_name',values='total_assets')


df3_corr = df3.pct_change().corr().fillna(0)
linkage = hc.linkage(df3_corr, method='average')
sns.clustermap(df3_corr, row_linkage=linkage, col_linkage=linkage,cmap="RdBu_r")
```
